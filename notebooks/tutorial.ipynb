```python
# Import the package
import visualize_pixel_map as vpm
```

```python
# Initialize Data with CSV file and custom time bins
# Note: Use start_time, end_time, and time_step in nanoseconds
data = vpm.Data(
    csv_filepath="data.empirphot.csv",
    start_time=0.0,
    end_time=0.01,
    time_step=10.0  # 10 ns step
)
```

```python
# Basic plot with zoom and colormap
data.plot(
    key=800,
    despine=False,
    zoom_size=25,
    cmap="viridis_r",
    time_bins=50,  # [10, 20, 30, 40, 50] ns
    show_scale=True
)
```

```python
# Plot with different start index and colormap
data.plot(
    key=590,
    despine=False,
    zoom_size=25,
    cmap="magma_r",
    time_bins=80,  # [10, 20, 30, 40, 50, 60, 70, 80] ns
    show_scale=True
)
```

```python
# Advanced plot with custom region, no labels, and legend
data.plot(
    key=635,
    despine=False,
    zoom_region=[60, 80, 175, 220],  # [x_min, x_max, y_min, y_max]
    cmap="magma_r",
    time_bins=60,  # [10, 20, 30, 40, 50, 60] ns
    show_scale=True,
    show_labels=False,
    show_legend=True
)
```
